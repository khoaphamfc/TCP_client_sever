```python
import matplotlib.pyplot as plt
  
with open('window_size_data.txt', 'r') as text_file:
    data = text_file.read().split('\n')
    
window_size_every_period = eval(data[0])    
print(type(window_size_every_period))
time_list = []
for i in range(len(window_size_every_period)):
    window_size_every_period[i] = int(window_size_every_period[i])
    time_list.append(600 * (i+1))
graph_x_values = time_list
graph_y_values = window_size_every_period 
plt.plot(graph_x_values, graph_y_values)
plt.xlabel('Time in seconds')
plt.ylabel('Size of sliding window')
plt.title('Graph of size of sliding window through every period of time')

plt.show()
```
